### - column에 하나라도 `NaN` 이 있는 컬럼을 드랍하자

### - `해제사유발생일`, `번지`, `본번`, `부번` 은 열은 이제 다시는 사용할 일이 없다. 이 열들을 drop 하자.

### - column 명을 변경하자

- 계약년월 -> yearmonth,
- 계약일 -> date,
- 시군구 -> address,
- 단지명 -> apt,
- 전용면적(㎡) -> area,
- 거래금액(만원) -> transaction,
- 층 -> floor,
- 건축년도 -> built,
- 도로명 -> road

```python
rename_map={ "계약년월":'yearmonth',
    "계약일": 'date',
    "시군구":"address",
    "단지명":"apt",
    "전용면적(㎡)":"area",
    "거래금액(만원)":"transaction",
    "층":"floor",
    "건축년도":"built",
    "도로명":"road"}
df.rename(rename_map,axis='columns',inplace=True)
df
```

### - `yearmonth`와 `date` 열을 합쳐서 하나로 만들고 pandas 의 시계열 자료형으로 바꾸자

```python
import datetime
```

```python
df["cdate"] = pd.to_datetime(df['yearmonth'].astype('str') + df['date'].astype('str'), format="%Y%m%d")
df
```

### - 계약날짜를 년, 월, 날짜로 나누어 새로운 column으로 추가

```python
df['year'] = df.cdate.dt.year
df['month'] = df.cdate.dt.month
df['day'] = df.cdate.dt.day
df
```

### - 주소를 각각 시, 구, 동 으로 나고 이를 column으로 저장

```python
df[['city', 'gu', 'dong']] = df.address.str.split(' ', n=2, expand=True)
df
```
